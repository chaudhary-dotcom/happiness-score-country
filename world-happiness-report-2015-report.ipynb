# Visualization
#### 1.1 Bar plot of top 10 Happiest country and top 10 Unhappiest country

```python
# Get the top 10 and bottom 10 countries by happiness score
top10 = df.nlargest(10, 'Happiness Score')
bottom10 = df.nsmallest(10, 'Happiness Score')

fig, axes = plt.subplots(1,2,figsize=(18,8))

# Top 10 countries
sns.barplot(ax=axes[0], y='Country', x='Happiness Score', data=top10.sort_values('Happiness Score', ascending=False), palette='viridis')
axes[0].set_xlabel('Happiness Score', fontsize=12)
axes[0].set_ylabel('Country', fontsize=12)
axes[0].set_title('Top 10 Happiest Countries (2015)', fontsize=15)

for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%.2f', padding=3)

# Bottom 10 countries
sns.barplot(ax=axes[1], y='Country', x='Happiness Score', data=bottom10.sort_values('Happiness Score'), palette='viridis')
axes[1].set_xlabel('Happiness Score', fontsize=12)
axes[1].set_ylabel('Country', fontsize=12)
axes[1].set_title('Bottom 10 Countries Unhappy Countries (2015)', fontsize=15)

for container in axes[1].containers:
    axes[1].bar_label(container, fmt='%.2f', padding=3)

plt.tight_layout()
plt.show()
```

#### 1.2 Happiest country by Region

```python
region_mean = df.groupby('Region')['Happiness Score'].mean().sort_values(ascending=False)
region_per = (region_mean/region_mean.sum())*100
plt.figure(figsize=(15,5))
ax = sns.barplot(y=region_per.index,x=region_per.values, palette='magma')
plt.xlabel('Happiness Score mean Values', fontsize=14)
plt.ylabel('Region', fontsize=14)
plt.title('Happiness Score by Region (2015)', fontsize=18)

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%', padding=3)

plt.tight_layout()
plt.show()
```

#### 1.3 Public Trust with the government according to Region (2015)

```python
trust_region = df.groupby('Region')['Trust (Government Corruption)'].mean().sort_values(ascending=False)
trust_region_per = (trust_region/trust_region.sum())*100

plt.figure(figsize=(14,8))
ax = sns.barplot(y=trust_region_per.index, x=trust_region_per.values, palette='deep')
plt.xlabel('Government trust (%)', fontsize=12)
plt.ylabel('Region', fontsize=12)
plt.title('Public Trust with Government', fontsize=16, color='r')
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%', padding=3)

plt.tight_layout()
plt.show()
```

#### 1.4 Highest to lowest Corrupt Country

```python
plt.figure(figsize=(20,80))
country_mean_gov = df.groupby('Country')['Trust (Government Corruption)'].mean().sort_values(ascending=False)
country_mean_gov_per = (country_mean_gov/country_mean_gov.sum())*100

ax = sns.barplot(y=country_mean_gov_per.index, x=country_mean_gov_per.values, palette='deep')
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%', padding=3)
    
    plt.title('Government trust', fontsize=18)
plt.tight_layout()
```

#### 1.3 Factors Contributing to Happiness in 2015

```python
# create scatterplots for each factors vs Happiness Score (2015)
factors = df[['Standard Error', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']]
fig, axes = plt.subplots(2,4, figsize=(20,12))
axes = axes.flatten()

for i, factor in enumerate(factors):
    sns.regplot(ax=axes[i],x=factor, y='Happiness Score', data=df, scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})
    axes[i].set_xlabel(factor)
    axes[i].set_ylabel('Happiness Score')
    axes[i].set_title(f'{factor} vs Happiness Score', fontsize=12)

    # add correlation coefficient 
    corr = df[factor].corr(df['Happiness Score'])
    axes[i].annotate(f'r={corr:.2f}', xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12, bbox=dict(boxstyle='round,pad=0.3', fc='white', ec='b', lw=1))
plt.tight_layout()
plt.suptitle('Relationship Between Factors and Happiness Score (2015)', fontsize=20, y=1.02, color='b')
    
plt.show()
```

```python
df_numeric = df.select_dtypes(include=['number'])
corr_matrix = df_numeric.corr()
corr_with_happiness = corr_matrix['Happiness Score'].sort_values(ascending=False)

sns.heatmap(corr_with_happiness.to_frame(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation with Happiness Score', color='b')
```
